## Data: Poverty Rate for Eastern Indonesia Provinces, 2010 - 2022
Data obtained from Badan Pusat Statistik


### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Loading and Understanding Data

```python
import pandas as pd

# Define the file path
file_path = '/content/drive/MyDrive/reg3/kemiskinan.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the DataFrame
df.head()
```

**Change the year format**

```python
df['Tahun'] = df['Tahun'].astype(str).str.extract('(\d{4})')
df['Tahun'] = pd.to_datetime(df['Tahun'], format='%Y')
df['Tahun'] = df['Tahun'].dt.year
df.set_index('Tahun', inplace=True)
```

```python
df.describe().T
```

**Checking for Null Data**

```python
df.isnull().sum()
```

### Data Analysis

```python
data_to_plot = ['Maluku Tenggara Barat', 'Maluku Tenggara', 'Maluku Tengah', 'Buru', 'Kepulauan Aru', 'Seram Bagian Barat', 'Seram Bagian Timur', 'Maluku Barat Daya', 'Buru Selatan', 'Ambon', 'Tual']

df[data_to_plot].plot(kind='line', marker='.', figsize=(6, 6))

# Customize the plot
plt.xlabel('Tahun')
plt.ylabel('Persentase (%)')
plt.title('Persentase Kemiskinan Provinsi Maluku')
plt.legend(title='Kabupaten/Kota')

# Set y-axis limits
plt.ylim(0, 100)

# Show the plot
plt.show()
```

According to the graph, it is evident that Ambon City has the lowest poverty rate among the cities in Maluku Provinces. Overall, Maluku Provinces experienced a decline in the poverty rate from 2010 to 2022.
